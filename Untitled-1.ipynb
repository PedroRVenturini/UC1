```python
temp = []

for i in range(6):
    Temperatura = float(input('Temperatura: '))
    temp.append(Temperatura)


for i in temp:
    if i < 15:
        print('Frio')
    elif i >= 15 and i <= 28:
        print('Agradavel')
    else:
        print('Quente')
```

```python
nomes = []
precos = []

for i in range(2):
    nome = input('Nome do produto: ')
    preco = float(input('Preço do produto: '))
    nomes.append(nome)
    precos.append(preco)

total = 0

for preco in precos:
    total = total + preco
    print(total)

```
