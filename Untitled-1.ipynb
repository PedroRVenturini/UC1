```python
sair = 'SIM'
candidatos = {
    'nomes':[],
    'idades':[],
    'cidades':[],
    'vagas':[]
}

while sair == 'SIM':
    nome = input('Nome: ')
    idade = int(input('Idade: '))
    cidade = input('Cidade: ')
    vaga = input('Vagas: ')

    candidatos['nomes'].append(nome)
    candidatos['idades'].append(idade)
    candidatos['cidades'].append(cidade)
    candidatos['vagas'].append(vaga)
    sair = input('Desejo cadastrar mais candidatos?').strip().upper()


for i in range(len(candidatos['nomes'])):
    print(f'O candidato {candidatos['nomes'][i]} tem {candidatos['idades'][i]}anos, mora na cidade de {candidatos['cidades'][i]} candidato a vaga {candidatos['vagas'][i]}')
```

```python
while True:
    try:
        num1 = float(input("Digite o primeiro número: "))
        operacao = input("Digite a operação (+, -, *, /): ")
        num2 = float(input("Digite o segundo número: "))

        if operacao not in ("+", "-", "*", "/"):
            print("Erro: operação inválida. Use apenas +, -, * ou /.")
            continue

    except ValueError:
        print("Erro: digite um número válido.")
        continue

    else:
        try:
            if operacao == "+":
                resultado = num1 + num2
            elif operacao == "-":
                resultado = num1 - num2
            elif operacao == "*":
                resultado = num1 * num2
            elif operacao == "/":
                resultado = num1 / num2

            print(f"Resultado: {resultado}")
            break

        except ZeroDivisionError:
            print("Erro: não é possível dividir por zero.")
```
